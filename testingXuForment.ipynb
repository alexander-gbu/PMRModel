```python
import numpy as np

R = 8.314
T = 510 + 273.15
P = 16

X_CH4 = 0.22909205
X_H2O = 0.75960961
X_CO = 0
X_H2 = 0.00007212   
```

```python
A_CO = 8.23 * 10**-5
A_H2 = 6.12 * 10**-9
A_CH4 = 6.65 * 10**-4
A_H2O = 1.77 * 10**5

D_H_CO = -70.65 # KJ/mol
D_H_H2 = -82.9 # KJ/mol
D_H_CH4 = -38.28 # KJ/mol
D_H_H2O = 88.68 # KJ/mol

D_H_CO = D_H_CO * 1000 # J/mol
D_H_H2 = D_H_H2 * 1000 # J/mol
D_H_CH4 = D_H_CH4 * 1000 # J/mol
D_H_H2O = D_H_H2O * 1000 # J/mol

K_CO = A_CO * np.exp(-D_H_CO / (R * T)) # bar**-1
K_H2 = A_H2 * np.exp(-D_H_H2 / (R * T)) # bar**-1
K_CH4 = A_CH4 * np.exp(-D_H_CH4 / (R * T)) # bar**-1
K_H2O = A_H2O * np.exp(-D_H_H2O / (R * T)) # 1

print([K_CO,K_H2,K_CH4,K_H2O])

P_CH4 = X_CH4*P #bar
P_H2O = X_H2O*P #bar
P_CO = X_CO*P #bar
P_H2 = X_H2*P #bar

DEN = 1 + K_CO * P_CO + K_H2 * P_H2 + K_CH4 * P_CH4 + K_H2O * P_H2O / P_H2
print(DEN)

A1 = 4.225 * 10**15

E1 = 240.1 # KJ/mol
E1 = E1 * 1000 # J/mol

k1 = A1 * np.exp(-E1/(R*T)) # kmol*bar**(1/2) / (kgcat*hr)
K1 = 119.8 * 10**11 * np.exp(-26830/T) #bar2
print(k1)

rate1 = k1 / (P_H2**2.5) * (P_CH4 * P_H2O - P_H2**3 * P_CO/K1) / (DEN**2)
print(rate1*3.6*67)
print(K1)
```
