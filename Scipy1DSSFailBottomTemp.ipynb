```python
#1D non-isothermal

'''
 - I THINK IT IS BEST IF I MOVE FORWARD WITH ONE LONG PFR, THEN WE GO FROM THERE THE TEMP WILL BE CONFUSING BUT THEN THE CONCENTRATIONS AND THE TRANSITION ACROSS THE MIDDLE WILL BE BETTER.
 WILL HAVE TO USE X_NUM/2 BECAUES OTHERWISE RUNNING INTO TROUBLE!!!
'''

from scipy.integrate import solve_bvp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def heats(T):
    #heats of reaction and heat capacity calculation from NIST
    t = T/1000

    #CH4
    A =	-0.703029
    B =	108.4773
    C =	-42.52157
    D =	5.862788
    E =	0.678565
    F =	-76.84376
    G =	158.7163
    H =	-74.87310

    dH_CH4 = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
    Cp_CH4 = A + B*t + C*t**2 + D*t**3 + E/(t**2)

    #H2O
    A =   30.09200
    B =   6.832514
    C =   6.793435
    D =  -2.534480
    E =   0.082139
    F =  -250.8810
    G =   223.3967
    H =  -241.8264

    dH_H2O = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
    Cp_H2O = A + B*t + C*t**2 + D*t**3 + E/(t**2)

    #CO
    A =   25.56759
    B =   6.096130
    C =   4.054656
    D =  -2.671301
    E =   0.131021
    F =  -118.0089
    G =   227.3665
    H = -110.5271

    dH_CO = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
    Cp_CO = A + B*t + C*t**2 + D*t**3 + E/(t**2)

    #CO2
    A =   24.99735
    B =   55.18696
    C =  -33.69137
    D =   7.948387
    E =  -0.136638
    F =  -403.6075
    G =   228.2431
    H =  -393.5224

    dH_CO2 = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
    Cp_CO2 = A + B*t + C*t**2 + D*t**3 + E/(t**2)

    #H2
    A =  33.066178
    B = -11.363417
    C =  11.432816
    D = -2.772874
    E = -0.158558
    F = -9.980797
    G =  172.707974
    H =  0.0

    dH_H2 = A*t + B*t**2/2 + C*t**3/3 + D*t**4/4 - E/t + F - H
    Cp_H2 = A + B*t + C*t**2 + D*t**3 + E/(t**2)

    #Ar
    A =	20.78600
    B =	2.825911*10**-7
    C =	-1.464191*10**-7
    D =	1.092131*10**-8
    E =	-3.661371*10**-8
    F =	-6.197350
    G =	179.9990
    H =	0

    Cp_Ar = A + B*t + C*t**2 + D*t**3 + E/(t**2) #J/mol K

    dHsmr = (3*dH_H2 + dH_CO - dH_CH4 - dH_H2O + 206.1)*10**3 #J/mol
    dHwgs = (dH_H2 + dH_CO2 - dH_CO - dH_H2O - 41.15)*10**3 #J/mol
    # print(dHsmr)
    # print(dHwgs)
    # print([Cp_CH4, Cp_H2O, Cp_CO, Cp_CO2, Cp_H2])
    return dHsmr, dHwgs, [Cp_CH4, Cp_H2O, Cp_CO, Cp_CO2, Cp_H2, Cp_Ar]
```

```python
#reactor constants
L = 0.0609 #m
R_inner = 2.39/1000 #m
R_outer = 3.5/1000 #m

#reactor discritization
x_num = 1000
x_pts = np.linspace(0,L,x_num)

A = np.pi * ((R_outer)**2-(R_inner)**2) #m^2 AXIAL AREA (GAS FLOW)
V = A*L #m^3
Souter = 2*np.pi*R_outer*L #m^2 outer RADIAL AREA (HEAT FLUX AND REACTION)
Sinner = 2*np.pi*R_inner*L #m^2 inner RADIAL AREA (HEAT FLUX)

h2Removal = 0 #change to 1 if there is removal of hydrogen

#gas flow constants WHEN CHANGING THE TEMPERATURE YOU NEED TO ADJUST THE INITIAL GUESS -- USING ODEINT MIGHT BE FOR BETTER CONVERSION
Tingoal = 622+273 #K
P = 1 #bar or atm
Rgas = 8.3144598 * 10**-5

Tinnerin = 298

sccm0 = np.array([16.16, 48.48, 0, 0, 10, 5])
mols0 = sccm0/22400 * 60
y0 = mols0/sum(mols0)
Ctot = P/(Rgas * Tingoal)
C0 = y0*Ctot
C0 = np.concatenate((C0, [Tingoal], [Tinnerin]))
print(C0)

#this will be v0 on the inner tube
u0 = sum(sccm0) * Tinnerin / 273 * 1 / P / 60 #cm^3/sec
v0 = u0 / (100**3 * A) #m/sec

#rate equation constants
Rreact = 8.314 #J / molÂ·K
Easmr = 165.740 #kJ/mol
Asmr = 1.68*10**8
Eawgs = 89.23 #kJ/mol
Awgs = 9.90*10**3

Ku = 1

I = 9 #curent in Amps
F = 96485 #faradays constant

Uouter = 0.116/Souter #J (s K m2)^-1.
Ts = 298 #K
Uinner = Uouter #This we can change later

q_furn= 1000 #J/s

#velocity balance
MM = np.array([16.04, 18.01528, 28.01, 44.009, 2.016, 39.948])
m0 = v0 * A * np.sum(y0[:6]*MM) * P/(Rgas * Tinnerin)
print(m0) #kg/s
```

```python
#initial guess array
init_guess = np.zeros((8, x_num))
init_guess[0, :] = C0[0]
init_guess[1, :] = C0[1]
init_guess[2, :] = C0[2]
init_guess[3, :] = C0[3]
init_guess[4, :] = C0[4]  
init_guess[5, :] = C0[5]
init_guess[6, :] = 900
init_guess[7, :] = np.linspace(900, Tinnerin, x_num)

def odes(x, C): #dy/dt = f(x)
    global vfinal, Tin
    T = C[6]
    Tin = C[7,0]
    Tinner = C[7]
    
    Ctot = P/(Rgas * T)
    y = C[:6]/np.sum(C[:6], axis = 0)

    v = 1/(np.dot(MM.T,y) *Ctot * A) * m0   #v * A * np.sum(soln.y[:6,-1]*MM) = m0
    vfinal = v[-1]

    Keqsmr = (101325/100000)**2 * np.exp(-26830/T + 30.114)
    Keqwgs = np.exp(4400/T - 4.036)

    rsmr = Souter/V*Ku*Asmr*np.exp(-Easmr*1000/Rreact/T)*(y[0]*y[1]-(P**2)*y[2]*(y[4]**3)/Keqsmr)
    rwgs = Souter/V*Ku*Awgs*np.exp(-Eawgs*1000/Rreact/T)*(y[1]*y[2]-y[3]*y[4]/Keqwgs)
    rh2r = I/(2*F*L) #have not confirmed if this is correct

    dHsmr, dHwgs, Cps = heats(T)

    Cpg = np.sum(y[:6, :]*Cps, axis=0)

    Tempinner = Sinner*Uinner*(T[:] - Tinner)/(v*Ctot*Cpg*V)
    Temp = (Souter*Uouter*(-T[:] + Ts) + Sinner*Uinner*(-T[:] + Tinner) - V*(dHsmr*rsmr + dHwgs*rwgs) + q_furn)/(v*Ctot*Cpg*V)
    ch4 = (-rsmr)/v
    h2o = (-rsmr - rwgs)/v
    co = (rsmr - rwgs)/v
    co2 = (rwgs)/v
    h2 = (3*rsmr + rwgs + h2Removal*rh2r)/v
    ar = np.zeros(len(T))
    
    return np.vstack([ch4, h2o, co, co2, h2, ar, Temp, Tempinner])

def bcs(yleft, yright):
    Ctot = P/(Rgas * Tin)
    C0 = y0*Ctot
    C0 = np.concatenate((C0, [Tin], [Tinnerin]))
    return np.array([yleft[0]-C0[0], yleft[1]-C0[1], yleft[2]-C0[2], yleft[3]-C0[3], yleft[4]-C0[4], yleft[5]-C0[5], yleft[6]-C0[6], yright[7]-C0[7]])

soln = solve_bvp(odes, bcs, x_pts, init_guess, max_nodes=1000, tol = 0.000001)
```
